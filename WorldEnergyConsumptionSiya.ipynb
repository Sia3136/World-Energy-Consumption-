#LSTM forecast for India

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
country_name = "India"
year_cols = [str(y) for y in range(1965, 2024)]

data = df[df['Country']==country_name][year_cols].values.flatten()

sequence_length = 10
X_seq, y_seq = [], []

for i in range(len(data)-sequence_length):
    X_seq.append(data[i:i+sequence_length])
    y_seq.append(data[i+sequence_length])

X_seq = np.array(X_seq)
y_seq = np.array(y_seq).reshape(-1, 1)
```

```python
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X_seq)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y_seq)

X_lstm = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)
```

```python
split = int(len(X_lstm)*0.8)
X_train, X_test = X_lstm[:split], X_lstm[split:]
y_train, y_test = y_scaled[:split], y_scaled[split:]
```

```python
model = Sequential()
model.add(LSTM(50, activation='tanh', input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=200, batch_size=1, verbose=0)
```

```python
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_actual = scaler_y.inverse_transform(y_test)
```

```python
mse = mean_squared_error(y_test_actual, y_pred)
mae = mean_absolute_error(y_test_actual, y_pred)
r2 = r2_score(y_test_actual, y_pred)

print(f"LSTM MSE: {mse:.3f}")
print(f"LSTM MAE: {mae:.3f}")
print(f"LSTM R²: {r2:.3f}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(len(y_test_actual)), y_test_actual, marker='o', label='Actual')
plt.plot(range(len(y_pred)), y_pred, marker='x', linestyle='--', label='LSTM Prediction')
plt.xlabel("Time Step")
plt.ylabel("Energy Consumption")
plt.title(f"LSTM Forecast for {country_name}")
plt.legend()
plt.grid(True)
plt.show()
```



> **LSTM predictions vs actual show how the sequence model performs on India’s test set. If LSTM produces unrealistic spikes or poor generalization, that suggests insufficient data volume or model complexity mismatch.**




#ARIMA Forecast for India

```python
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
country_name = "India"
country_data = df[df['Country'] == country_name]

year_cols = [col for col in df.columns if col.isdigit()]
country_series = country_data[year_cols].iloc[0].apply(pd.to_numeric, errors='coerce').dropna()
country_series.index = pd.to_datetime(country_series.index, format='%Y')
```

```python
arima_model = auto_arima(country_series, seasonal=False, stepwise=True, suppress_warnings=True)
print(arima_model.summary())

from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(country_series, order=arima_model.order)
model_fit = model.fit()
```

```python
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)
forecast[forecast < 0] = 0
```

```python
plt.figure(figsize=(12,6))
plt.plot(country_series.index.year, country_series.values, marker='o', label="Actual")
plt.plot(country_series.index.year, model_fit.fittedvalues.values, linestyle='--', label="Fitted")
forecast_years = list(range(country_series.index.year[-1]+1, country_series.index.year[-1]+forecast_steps+1))
plt.plot(forecast_years, forecast.values, marker='o', linestyle='--', color='red', label="Forecast")
plt.xlabel("Year")
plt.ylabel("Energy Consumption")
plt.title(f"ARIMA Forecast for {country_name}")
plt.legend()
plt.grid(True)
plt.show()
```



> **ARIMA was fitted and a 10-year forecast plotted alongside fitted values and observed series.In this dataset ARIMA produced smoother, more plausible forecasts than LSTM and the linear approach.**



```python
d = arima_model.order[1]
y_pred_in_sample = model_fit.fittedvalues
if d > 0:
    y_pred_in_sample = y_pred_in_sample.iloc[d:]
y_true = country_series.iloc[d:]

mse = mean_squared_error(y_true, y_pred_in_sample)
mae = mean_absolute_error(y_true, y_pred_in_sample)

print(f"In-sample MSE: {mse:.3f}")
print(f"In-sample MAE: {mae:.3f}")
```

```python
years_lstm_test = np.arange(2014, 2024)

plt.figure(figsize=(12,6))
plt.plot(years_lstm_test, y_test_actual.flatten(), marker='o', label='Actual (India)', color='black')
plt.plot(years_lstm_test, y_pred.flatten(), marker='x', linestyle='--', label='LSTM Prediction (India)')
india_df = df[df['Country'] == 'India']
india_features_X = india_df[X_years + ['Country_Code']]
india_scaled_X = scaler.transform(india_features_X)
india_pca_X = pca.transform(india_scaled_X)
india_ridge_predictions_full = ridge_model.predict(india_pca_X)
india_ridge_predictions_overlap = india_ridge_predictions_full[0, 3:]

plt.plot(years_lstm_test, india_ridge_predictions_overlap, marker='s', linestyle=':', label='Ridge Prediction (India)')

plt.xlabel("Year")
plt.ylabel("Energy Consumption")
plt.title("Comparison of Model Predictions for India (2014–2023)")
plt.legend()
plt.grid(True)
plt.show()
```

**ARIMA performed better than Linear Regression and LSTM because:**

1. Your data is time-series with clear yearly patterns

ARIMA is specifically designed for time-series forecasting, unlike Linear Regression (general-purpose) or LSTM (needs large data).

2. You have limited data (only ~50 years)

LSTM needs thousands of sequences to learn well.
ARIMA works extremely well even with small datasets, making it ideal here.

 3. ARIMA captures trends and seasonality effectively

Energy consumption usually changes gradually over years.
ARIMA models trend + difference patterns, which matches your data structure.

 4. Linear Regression overfitted

Your linear model had extremely high R² but poor generalization.
This means it was memorizing patterns instead of learning them.

5. ARIMA gave stable predictions without extreme spikes

LSTM sometimes gave unrealistic values because the dataset was small.
ARIMA gave smooth, realistic, and consistent forecasts, matching historical behavior.
